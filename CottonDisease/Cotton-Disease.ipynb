```python
train_ds  = ah.KroniaDataset(data=X_train,transforms=imgTransform)
val_ds = ah.KroniaDataset(data=X_train,transforms=imgTransform)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
class CottonTypeModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(3,16,3,1)
        self.conv2 = nn.Conv2d(16,32,3,1)
        self.conv3 = nn.Conv2d(32,64,3,1)
        self.conv4 = nn.Conv2d(64,64,3,1)
        self.conv5 = nn.Conv2d(64,64,3,1)
        self.fc1 = nn.Linear(1600,128)
        self.fc2 = nn.Linear(128,4)
    
    def forward(self,X):
        X = F.relu(self.conv1(X))
        X = F.max_pool2d(X,2,2)
        
        X = F.relu(self.conv2(X))
        X = F.max_pool2d(X,2,2)
        
        X = F.relu(self.conv3(X))
        X = F.max_pool2d(X,2,2)
        
        X = F.relu(self.conv4(X))
        X = F.max_pool2d(X,2,2)
        
        X = F.relu(self.conv5(X))
        X = F.max_pool2d(X,2,2)
        
        X = X.view(-1,1600)
        
        X = F.relu(self.fc1(X))
        X = self.fc2(X)
        return F.log_softmax(X,dim=1)
```

```python
torch.manual_seed(42)
model = CottonTypeModel()
```

```python
model
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
from torch.utils.data import DataLoader
```

```python
train_dl = DataLoader(dataset=train_ds,batch_size=32,shuffle=True)
val_dl = DataLoader(dataset=val_ds,batch_size=32,shuffle=True)
```

```python
train_loss,val_loss = mch.trainModel(model=model,train_dl=train_dl,val_dl=val_dl,criterion=criterion,optim=optimizer,train_samples=5770,batch_size=32,soft_max=False)
```

```python
mch.visualiziseTrainResults(train_losses=train_loss,test_losses=val_loss)
```
